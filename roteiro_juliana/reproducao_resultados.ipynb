```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lti, step

# Parâmetros do sistema
K_process = 1.7
tau_process = 45.0

# Parâmetros do controlador PI
tau_c_reg = 10.0  # Constante de tempo desejada para controle regulatório
tau_c_servo = 25.0  # Constante de tempo desejada para controle servo

# Ganho do controlador PI
Kc_reg = tau_process / (K_process * tau_c_reg)
Kc_servo = tau_process / (K_process * tau_c_servo)

# Constante de tempo integrativa do controlador PI
tau_i = tau_process

# Definindo a função de transferência do processo
num = [K_process]
den = [tau_process, 1]
process_system = lti(num, den)

# Simulação de resposta ao degrau
t = np.linspace(0, 200, num=1000)
t_reg, y_reg = step(process_system, T=t)
t_servo, y_servo = step(process_system, T=t)

# Plotagem das respostas
plt.figure(figsize=(10, 6))
plt.plot(t_reg, y_reg, label='Controle Regulatório (Reg)')
plt.plot(t_servo, y_servo, label='Controle Servo')
plt.xlabel('Tempo (s)')
plt.ylabel('Resposta do Sistema')
plt.title('Resposta do Sistema de Controle')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leia o arquivo CSV
df = pd.read_csv('CORTELETTIDIST.CSV', sep=';', parse_dates=[['$Date', '$Time']], dayfirst=True)

# Renomeie a coluna combinada de data e hora
df.rename(columns={'$Date_$Time': 'DateTime'}, inplace=True)

# Defina a coluna DateTime como índice
df.set_index('DateTime', inplace=True)

# Converta as colunas de dados para numérico, se necessário
df['VP'] = pd.to_numeric(df['VP'], errors='coerce')
df['SP'] = pd.to_numeric(df['SP'], errors='coerce')
df['SC'] = pd.to_numeric(df['SC'], errors='coerce')

# Plotar os dados
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['VP'], label='VP')
plt.plot(df.index, df['SP'], label='SP')
plt.plot(df.index, df['SC'], label='SC')
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Gráfico dos Sinais VP, SP e SC')
plt.legend()
plt.show()
```
