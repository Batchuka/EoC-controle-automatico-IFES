## <font color='green' style='font-size: 22px;'> Adimitindo uma Gc(s)=k, esboce o lugar das raízes e use a condição de módulo para determinar a faixa de valores que k pode assumir para resultar em uma malha fechada com resposta oscilatória </font> 

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl

# parâmetro k
k = 1

# Definindo a função de transferência da malha aberta 
G = ctl.tf([k*1], [1, 10, -100])

# Defina o tamanho do gráfico e a densidade de pontos
plt.figure(figsize=(10,10), dpi=80)

# Plote o lugar das raízes
ctl.root_locus(G, plot=True, grid=True, print_gain=True)

# Configurações do gráfico
plt.title('Lugar das Raízes')
plt.xlabel('Eixo Real')
plt.ylabel('Eixo Imaginário')
plt.grid(True)

# Exiba o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl

# Valor de 'k' calculado.
k = 1

# Definindo a função de transferência de 3ª ordem
num1 = [k]
den1 = [1, -90, k]
sys1 = ctl.TransferFunction(num1, den1)

# Obtendo a resposta ao degrau
t, y1 = ctl.step_response(sys1, T=np.linspace(0, 10, 1000), input=0)

# Plotando o gráfico
plt.plot(t, y1, label='2ª ordem')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída')
plt.title('Resposta ao degrau para k={}'.format(k))
plt.grid()
plt.legend()
plt.show()
```
