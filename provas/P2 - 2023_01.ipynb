## <font color='green' style='font-size: 22px;'> Adimitindo uma Gc(s)=k, esboce o lugar das razes e use a condio de mdulo para determinar a faixa de valores que k pode assumir para resultar em uma malha fechada com resposta oscilatria </font> 

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl

# parmetro k
k = 1

# Definindo a funo de transferncia da malha aberta 
G = ctl.tf([k*1], [1, 10, -100])

# Defina o tamanho do grfico e a densidade de pontos
plt.figure(figsize=(10,10), dpi=80)

# Plote o lugar das razes
ctl.root_locus(G, plot=True, grid=True, print_gain=True)

# Configuraes do grfico
plt.title('Lugar das Razes')
plt.xlabel('Eixo Real')
plt.ylabel('Eixo Imaginrio')
plt.grid(True)

# Exiba o grfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl

# Valor de 'k' calculado.
k = 1

# Definindo a funo de transferncia de 3 ordem
num1 = [k]
den1 = [1, -90, k]
sys1 = ctl.TransferFunction(num1, den1)

# Obtendo a resposta ao degrau
t, y1 = ctl.step_response(sys1, T=np.linspace(0, 10, 1000), input=0)

# Plotando o grfico
plt.plot(t, y1, label='2 ordem')
plt.xlabel('Tempo (s)')
plt.ylabel('Sada')
plt.title('Resposta ao degrau para k={}'.format(k))
plt.grid()
plt.legend()
plt.show()
```
